```python
print("List Data Structure")
```

```python
# Empty list
lst = []

# With elements
lst = [1, 2, 3, 4]

# Mixed data types
lst = [1, "hello", 3.5, True]

# Using constructor
lst = list((10, 20, 30))
```

```python
lst = [10, 20, 30, 40]

print(lst[0])     # 10
print(lst[-1])    # 40 (last element)
```

```python
# Slicing
lst = [0, 1, 2, 3, 4, 5]

print(lst[1:4])   # [1,2,3]
print(lst[:3])    # [0,1,2]
print(lst[::2])   # [0,2,4]
print(lst[::-1])  # Reverse list
```

```python
# Modifying List

lst = [1, 2, 3]

lst[0] = 100   # Change value
```

```python
# Adding elements 

lst = [1, 2]

lst.append(3)          # [1,2,3]
lst.insert(1, 10)      # [1,10,2,3]

lst.extend([4, 5])     # [1,10,2,3,4,5]
```

```python
# Removing Elements

lst = [1, 2, 3, 4]

lst.remove(2)     # removes value
lst.pop()         # removes last
lst.pop(1)        # removes index

del lst[0]        # delete by index
lst.clear()       # empty list
```

```python
# Looping through list 

lst = [1 , 2 , 3 , 4]

for x in lst:
    print(x)

# With index
for i in range(len(lst)):
    print(lst[i])

# Pythonic way
for i, val in enumerate(lst):
    print(i, val)
```

```python
# Important Functions 

lst = [3, 1, 4, 2]

len(lst)        # 4
max(lst)        # 4
min(lst)        # 1
sum(lst)        # 10
```

```python
# Sorting 

lst.sort()              # ascending
lst.sort(reverse=True)  # descending

sorted_lst = sorted(lst)  # new list
```

```python
# Searching 

lst = [10, 20, 30]

print(20 in lst)     # True
print(lst.index(30)) # 2
```

```python
# Copying list 

lst1 = [1, 2, 3]

lst2 = lst1        # ❌ same reference
lst3 = lst1.copy() # ✅ new copy
```

```python
# List Comprehension

# Normal way
lst = []
for i in range(5):
    lst.append(i)

# Short way
lst = [i for i in range(5)]

# With Condition

lst = [i for i in range(10) if i % 2 == 0]
```

```python
# Nested Lists 

matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

print(matrix[0][1])  # 2
```

```python
# Common Tricks 

# 1. Reverse List 
lst[::-1]

# 2. Remove Duplicates 
lst_with_dupes = [1, 2, 2, 3, 3, 3, 4]
lst_no_dupes = list(set(lst_with_dupes))

# 3. Count frequency
lst.count(2)

# 4. Merge Lists 
lst3 = lst1 + lst2

# 5. Flatten list 
lst_mat = [
    [1, 2, 3],
    [4, 5, 6]
]

flat = [item for sublist in lst_mat for item in sublist]
print(flat)
```
