```python
a = [2 , 6 , 5, 8 , 11]
target = 14
```

**Brute Force - O(N)**

```python
def two_sum(a , target):
    for i in range(len(a)):
        for j in range(i + 1 ,len(a)):
            if a[i] + a[j] == target:
                return i , j 
    return -1 , -1
```

**Optimal Solution - O(N) , Extra space - O(N)**

```python
def two_sum(a , target):
    track_sum = {}
    for i in range(len(a)):
        number = a[i]
        need = target - number 
        if need in track_sum:
            return track_sum[need] , i 
        track_sum[number] = i 
    return -1 , -1;
```

**Better Solution - O(N * logN) , Extra Space - O(N)**

```python
def two_sum(a , target):
    arr = [(num , i) for i , num in enumerate(a)]
    arr.sort()
    i = 0 
    j = len(arr) - 1
    while i < j :
        if arr[i][0] + arr[j][0] > target:
            j -= 1
        elif arr[i][0] + arr[j][0] < target:
            i += 1
        else:
            return arr[i][1] , arr[j][1]
    return -1 , -1
```
