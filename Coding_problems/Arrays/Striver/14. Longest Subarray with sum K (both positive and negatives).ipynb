```python
a = [1 , 2 , 3 , 1 , 1 , 1 , 1 , 4 , 2 , 3]
k = 3 
```

**Brute Force - O(N^3)**

Generate all subarrays

```python
def longest_subarray_sum(a , k):
    n = len(a)
    max_len = -1

    for i in range(n):
        for j in range(i , n):
            sum = 0
            for k in range(i , j + 1):
                sum += a[k]
            if sum == k:
                max_len = max(max_len , j - i + 1)

    return max_len
```

**Improved Brute Force - O(N^2)**

**Optimal Solution - O(N) , space extra - O(N)**

Case 1 : Both positive and negative
    Optimal solution for this case 

```python
def longest_subarray_sum(a , k):
    n = len(a)
    
    d = {}
    cur_sum = max_len = 0
    for i , val in enumerate(a):
        cur_sum += val 

        if cur_sum == k:
            max_len = max(max_len , i + 1)

        rem = cur_sum - k
        if rem in d:
            length = i - d[rem]
            max_len = max(max_len , length)

        if cur_sum not in d:
            d[cur_sum] = i

    return max_len
```

**Optimal Solution - O(N) , Space - O(1)**

Case 2 : Only positive
    - Optimal for this case

```python
def longest_subarray_sum(a , k):
    n = len(a)
    
    left = right = 0
    sum = 0

    max_len = 0
    
    while right < n:

        sum += a[right]

        while sum > k :
            sum -= a[left]
            left += 1

        if sum == k :
            max_len = max(max_len , right - left + 1)

        right += 1
    
    return max_len
```
