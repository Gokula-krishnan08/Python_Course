**Brute Force - O(N * N)**

```python
arr = [1 , 2 , 3 , 5 , 6, 7]
n = len(arr)


for i in range(1 , n + 1):
    flag = 0
    for j in arr:
        if j == i:
            flag = 1
            break
    if flag == 0:
        print(f"Missing element : {i}")
```

**Better Solution - O(N) + O(N) , Extra Space - O(N)**

```python
arr = [1 , 2 , 3 , 5 , 6, 7]
n = 7   # range of elements : 1 - n

hash_array = [0] * (n + 1)

for num in arr:
    hash_array[num] += 1

for i in range(1 , n + 1):
    if hash_array[i] == 0:
        print(f"Misssing element : {i}")
        break
```

**Optimal Solution - O(N)**

Approach 1 : Sum of elements

```python
arr = [1 , 2 , 3 , 5 , 6, 7]
n = 7   # range of elements : 1 - n

sum = (n * (n + 1)) / 2
for num in arr:
    sum -= num 

print(f"Missing Element : {sum}")
```

**Optimal Solution - O(N)**

Approach 2 : XOR of elements

```python
arr = [1 , 2 , 3 , 5 , 6, 7]
n = 7   # range of elements : 1 - n

xor_1 = 0
for i in range(1 , n + 1):
    xor_1 ^= i

xor_2 = 0
for num in arr:
    xor_2 ^= num 

print(f"Missing Element : {xor_1 ^ xor_2}")
```
