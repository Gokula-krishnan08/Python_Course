**Brute Force - O(2 * N) , Extra space - O(N)**

```python
a = [1 , 0 , 2 , 3 , 2 , 0 , 0 , 4 , 5 , 1]

n = len(a)

temp = []
for num in a:
    if num != 0:
        temp.append(num)

a.clear()
for num in temp:
    a.append(num)

length = len(a)
for i in range( length , n):
    a.append(0)

print(a)
```

**Optimal Solution - O(N)**

```python
a = [1 , 0 , 2 , 3 , 2 , 0 , 0 , 4 , 5 , 1]

n = len(a)

i = -1
for j in range(n):
    if a[j] == 0:
        i = j
        break 

for j in range(i + 1 , n):
    if a[j] != 0:
        a[i] , a[j] = a[j] , a[i]
        i += 1



print(a)
```
