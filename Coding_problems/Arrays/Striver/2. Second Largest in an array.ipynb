**Brute Force - O(N logN)**

```python
a = [8 , 4 , 3 , 5]

a.sort()
first_largest = a[-1]
second_largest = a[-1]

for num in range(len(a) - 2 , -1 , -1):
    if a[num] != first_largest:
        print(a[num])
        second_largest = a[num]
        break

if second_largest == first_largest:
    print("No second largest element found.")
```

**Better Solution - O(2N)**

```python
a = [1 , 2 , 4 , 8 , 8]

first_largest = a[0]
for num in a:
    if num > first_largest:
        first_largest = num

second_largest = -1
for num in a:
    if num > second_largest and num != first_largest:
        second_largest = num 

print(second_largest)
```

**Optimal Solution - O(N)**

```python
a = [1 , 3 , 2 , 8 , 5]

first_largest = second_largest = -1

for num in a:
    if first_largest < num:
        second_largest = first_largest
        first_largest = num 
    elif second_largest < num and num != first_largest:
        second_largest = num 

print(second_largest)
```
