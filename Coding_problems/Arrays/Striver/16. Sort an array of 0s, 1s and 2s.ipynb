**Brute Force - O(N * logN)**

```python
def sort_an_array_of_012(a):
    a.sort()
    return a
```

**Better Solution - O(2*N) , Extra Space - O(1)**

```python
def sort_an_array_of_012(a):
    track = [0] * 3
    for num in a:
        track[num] += 1

    a.clear()
    for i , num in enumerate(track):
        for _ in range(num):
            a.append(i)

    return a 
```

**Optimal Solution - O(N) , Extra Space - O(1)**

Dutch National Flag algorithm

```python
def sort_an_array_of_012(a):
    n = len(a)
    i = j = 0
    k = n - 1

    while j <= k:
        if a[j] == 0:
            a[i] , a[j] = a[j] , a[i]
            i += 1
            j += 1
        elif a[j] == 1:
            j += 1
        else :
            a[k] , a[j] = a[j] , a[k]
            k -= 1
    
    return a
```
