**Brute Force - O(N)**

```python
arr = [1 , 1 , 2 , 3 , 3 , 4 , 4]
n = len(arr)

for num in arr:
    count = 0
    for j in range(n):
        if arr[j] == num:
            count += 1
    if count == 1:
        print(f"Element : {num}")
```

**Better Solution - O(N) , Extra space - O(N / 2)**


Infeasible if any element value is large , 
we can implement dictionary , instead of array to reduce unwanted array size

```python
arr = [1 , 2 , 2 , 3 , 3 , 4 , 4]
n = len(arr)

maximum = max(arr)
hash = [0] * (maximum + 1)

for num in arr:
    hash[num] += 1

for i , val in enumerate(hash):
    if val == 1:
        print(f"Element : {i}")
```

**Optimal Solution - O(N)**

XOR approach

```python
arr = [1 , 2 , 2 , 3 , 3 , 4 , 4]
n = len(arr)

xor = 0
for num in arr:
    xor ^= num 

print(f"Element : {xor}")
```
