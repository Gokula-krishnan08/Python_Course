**Brute Force - O(N1 * N2) , Extra Space - O(N1) or O(N2)**

```python
a = [1 , 2 , 3 , 3 , 4 , 4 , 6 , 8 , 9]
b = [1 , 3 , 3 , 5 , 8 , 9]

n = len(a)
m = len(b)

intersection = []
visited = [0] * m 

for i in range(n):
    for j in range(m):
        if a[i] == b[j] and visited[j] == 0:
            intersection.append(a[i])
            visited[j] = 1;
            break

        if b[j] > a[i]:
            break

print(intersection)
```

**Optimal Solution - O(N1 + N2) - only applicable if both arrays are sorted**

```python
a = [1 , 2 , 3 , 3 , 4 , 4 , 6 , 8 , 9]
b = [1 , 3 , 3 , 5 , 8 , 9]

n = len(a)
m = len(b)

i = j = 0
intersection = []

while i < n and j < m:
    if a[i] < b[j]:
        i += 1
    elif a[i] > b[j]:
        j += 1
    else:
        intersection.append(a[i])
        i += 1
        j += 1

print(intersection)
```
