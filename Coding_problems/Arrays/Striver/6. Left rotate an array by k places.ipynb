**Brute Force - O(N + k) , Extra space - O(k)**

```python
a = [1 , 2 , 3 , 4 , 5]

k = 3
k = k % len(a)

temp = []
for i in range(k):
    temp.append(a[i])

for i in range(k , len(a)):
    a[i - k] = a[i]

for i in range(len(a) - k , len(a)):
    a[i] = temp[i - (len(a) - k)]

print(a)
```

**Optimal Solution - O(N)**

```python
a = [1 , 2 , 3 , 4 , 5]

n = len(a)
k = 3 

k = k % n

temp_1 = a[0 : k]
temp_2 = a[k : n]

temp_1.reverse()
temp_2.reverse()

a = temp_1 + temp_2 
a.reverse()

print(a)
```
