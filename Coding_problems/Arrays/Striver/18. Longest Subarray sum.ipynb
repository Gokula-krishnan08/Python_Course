```python
a = [-2 , -3 , 4 , -1 , -2 , 1 , 5 , -3]
```

**Brute Force - O(N ^ 3)**

```python
def longest_subarray_sum(a):
    n = len(a)
    max_sum = 0

    for i in range(n):
        for j in range(i , n):
            summ = 0
            for k in range(i , j + 1):
                summ += a[k]

            max_sum = max(max_sum , summ);
    return max_sum            
```

**Better Solution - O(N ^ 2)**

```python
def longest_subarray_sum(a):
    n = len(a)
    max_sum = 0

    for i in range(n):
        summ = 0
        for j in range(i , n):

            summ += a[j]
            max_sum = max(max_sum , summ);
    
    return max_sum            
```

**Optimal solution - O(N)**

Kadane's Algorithm

```python
def longest_subarray_sum(a):
    n = len(a)
    max_sum = float('-inf')

    summ = 0 
    ans_start = ans_end = 0

    summ = 0
    start = 0

    for i , num in enumerate(a):
        if summ == 0:
            start = i

        summ += num 
        
        if summ > max_sum :
            max_sum = summ
            ans_start = start
            ans_end = i

        if summ < 0:
            summ = 0
            start = i + 1

    return ans_start , ans_end
```
