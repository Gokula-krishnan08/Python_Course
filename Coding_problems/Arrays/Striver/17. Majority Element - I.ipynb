```python
a = [2 , 2 , 3 , 3 , 1 , 2 , 2]
```

**Brute Force - O(N ^ 2)**

```python
def majority_element_I(a):
    for i in range(len(a)):
        count = 0
        for j in range(len(a)):
            if a[i] == a[j] :
                count += 1
        if count > (len(a) // 2) :
            return a[i]

    return -1
```

**Better Solution - O(N) , Extra Space - O(N)**

```python
def majority_element_I(a):
    track = {}
    for num in a:
        if num not in track:
            track[num] = 1
        else:
            track[num] += 1
    
    for i , val in track.items():
        if val > (len(a) // 2):
            return i
    
    return -1
```

**Optimal Solution - O(N) , Extra Space - O(1)**

```python
def majority_element_I(a):
    element = -1
    count = 0

    for num in a:
        if count == 0 :
            element = num 
            count += 1
        elif element == num :
            count += 1
        else :
            count -= 1

    count = 0
    for num in a:
        if num == element :
            count += 1

    if count > (len(a) // 2):
        return element 
    
    return -1
```
