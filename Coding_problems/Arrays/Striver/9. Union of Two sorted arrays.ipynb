**Brute Force - O(N1 + N2) + O(N1 + N2) , Extra Space - O(N1 + N2) + O(N1 + N2)**

```python
a = [1 , 1 , 2 , 3 , 4 , 5]
b = [2 , 3 , 4 , 4 , 5]

s = set()
for num in a:
    s.add(num)
for num in b:
    s.add(num)

union = []
for num in s:
    union.append(num)

print(union)
```

**Optimal Solution - O(N1 + N2) , Space complexity - O(N1 + N2)**

```python
a = [1 , 1 , 2 , 3 , 4 , 5]
b = [2 , 3 , 4 , 4 , 5]

i = j = 0

n = len(a)
m = len(b)

union = []

while i < n and j < m:
    if a[i] <= b[j]:
        if len(union) == 0 or union[-1] != a[i]:
            union.append(a[i])
        i += 1
    else:
        if len(union) == 0 or union[-1] != b[j]:
            union.append(b[j])
        j += 1
    
while i < n:
    if len(union) == 0 or union[-1] != a[i]:
        union.append(a[i])
    i += 1

while j < m:
    if len(union) == 0 or union[-1] != b[j]:
        union.append(b[j])
    j += 1

print(union)   
```
