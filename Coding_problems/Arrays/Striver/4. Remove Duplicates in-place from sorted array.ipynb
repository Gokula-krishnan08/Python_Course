**Brute Force - O(N logN)**

```python

a = [1 , 1 , 2 , 2 , 2 , 8 , 8 , 9]
s = set()
length = len(a)

for num in a:
    s.add(num)

a.clear()

for num in s:
    a.append(num)
a.sort()
for _ in range(length - len(a) , 0 , -1):
    a.append(0)

print(a)
```

**Optimal Solution - O(N)**

```python
a = [1 , 2 , 2 , 3 , 3 ,3 , 4]

i = 0
for j in range(1 , len(a)):
    if a[i] != a[j]:
        a[i + 1] = a[j]
        i += 1
    
print(i + 1) # length of the non-duplicated array
```
